### looking only at genes in both, how different are the expressions for each cell type

```python
# make sure column order is identical
assert lm22_real.shape[1] == lm22_my.shape[1]
lm22_my = lm22_my[lm22_real.columns]
assert all(lm22_my.columns == lm22_real.columns)
```

```python
genes_in_both = in_both.index
means=[]
# sanity check of how i'm indexing
assert lm22_real[lm22_real.index.isin(genes_in_both)].shape[0] == lm22_my[lm22_my.index.isin(genes_in_both)].shape[0]

# gen data
for c in lm22_real.columns:
    print c
    # get mean difference in expressions for each cell type, filtered to genes in both
    means.append((lm22_real[lm22_real.index.isin(genes_in_both)][c] - lm22_my[lm22_my.index.isin(genes_in_both)][c]).\
    mean())
means
```

```python
plt.hist(means)
```

Looks like nearly identical for all! (Note that this is not log transformed expression data, as far as I know)

## References (input to lm22 construction) look pretty consistent within themselves

```python
from sklearn.preprocessing import scale
# center data for each gene by mean, and set to unit variance
# set to unit variance
def normalize_mat(mat):
    #return scale(np.log2(mat))
    return np.log2(mat)
```

```python
for _,row in class_labels.iterrows():
    #print row.name
    reference_indices = (row[row==1]).index
    #print reference_indices
    # note that reference is not 1-indexed, because we did a setindex
    # so we have to subtract 1 from each column index
    matching = reference[reference.columns[reference_indices-1]] 
    #print matching.columns.tolist()
    new_mat = matching
    #signature = lm22_real[[row.name]] # keep it as a dataframe
    #new_mat = pd.merge(matching, signature, how='left', left_index=True, right_index=True) # last col is signature
    hm = pdh.DendroHeatMap(heat_map_data=normalize_mat(new_mat.as_matrix()))
    hm.row_labels = matching.index.tolist()
    hm.column_labels = matching.columns.tolist()
    hm.title=row.name
    hm.show()
    #break
```
