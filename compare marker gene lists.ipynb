```python
# %load ~/ipyhead
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
# from iris.ipynb
iris = pd.read_csv('iris_genes_by_category.csv')
del iris['Unnamed: 0']
iris.head()
```

```python
bindea = pd.read_csv('Bindea Immunome - Bindea paper.csv')
bindea.head()
```

```python
bindea.CellType.value_counts()
```

```python
genes_by_category = pd.DataFrame(bindea.groupby('CellType').AffymetrixID.unique()).transpose()
genes_by_category.head()
```

```python
bindea_genes = pd.DataFrame([genes_by_category[c].iloc[0] for c in genes_by_category]).transpose()
bindea_genes.columns = genes_by_category.columns
bindea_genes.head()
```

```python
iris['B Cell']
```

```python
iris.columns, bindea_genes.columns
```

```python
def compare_stats(left, right):
    missing_from_right = set(left) - set(right)
    missing_from_left = set(right) - set(left)
    intersection = set(left) & set(right)
    print 'Intersection: ', len(intersection)
    print 'Missing from left :', len(missing_from_left)
    print 'Missing from right :', len(missing_from_right)
    return intersection, missing_from_left, missing_from_right
```

```python
compare_stats(iris['B Cell'], bindea_genes['B cells'])
```

```python
compare_stats(iris['T Cell'], bindea_genes['T cells'])
```

```python
compare_stats(iris['NK Cell'], bindea_genes['NK cells'])
```
