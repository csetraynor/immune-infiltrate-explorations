```python
from scipy.stats import cauchy
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots()
def plot_cauchy(mu, gamma, ax):
    ax.plot(x, cauchy(mu, gamma).pdf(x), label='mu=%0.1f, gamma=%0.1f' % (mu, gamma))
plot_cauchy(0,2.5, ax)
plot_cauchy(0,1.5, ax)
plot_cauchy(0,.25, ax)
plt.legend()
```

```python
from scipy.stats import norm
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots()
def plot_norm(mu, sigma, ax):
    ax.plot(x, norm(mu, sigma).pdf(x), label='mu=%0.1f, sigma=%0.1f' % (mu, sigma))
plot_norm(0,1, ax)
plot_norm(0,0.5, ax)
plt.legend()
```
